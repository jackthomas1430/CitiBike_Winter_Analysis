```python
# Calculate the total number of rows in each dataset
total_rows_jan = len(jan_data)
total_rows_feb = len(feb_data)

# Calculate the number of rows with missing data in each dataset
missing_rows_jan = jan_data[jan_data.isnull().any(axis=1)].shape[0]
missing_rows_feb = feb_data[feb_data.isnull().any(axis=1)].shape[0]

# Calculate the percentage of missing data
percent_missing_jan = (missing_rows_jan / total_rows_jan) * 100
percent_missing_feb = (missing_rows_feb / total_rows_feb) * 100

print(f"Percentage of rows with missing data in January dataset: {percent_missing_jan:.2f}%")
print(f"Percentage of rows with missing data in February dataset: {percent_missing_feb:.2f}%")
```

```python
# Remove all rows with missing values 
## Remove rows with missing values in January dataset
jan_data_cleaned = jan_data.dropna()

## Remove rows with missing values in February dataset
feb_data_cleaned = feb_data.dropna()
```

```python
# Identify duplicate rows 
jan_data.drop_duplicates(inplace=True)
feb_data.drop_duplicates(inplace=True)
```

```python
# Check new row counts 

print(f"Original January dataset rows: {len(jan_data)}")
print(f"Cleaned January dataset rows: {len(jan_data_cleaned)}")
print(f"Original February dataset rows: {len(feb_data)}")
print(f"Cleaned February dataset rows: {len(feb_data_cleaned)}")
```

```python
# Standarize Station Coordinates 
station_coords = jan_data.groupby('start_station_name')[['start_lat', 'start_lng']].mean().reset_index()
jan_data = jan_data.merge(station_coords, on='start_station_name', suffixes=('', '_standardized'))
```

```python
# Repeat for feb_data 
feb_station_coords = feb_data.groupby('start_station_name')[['start_lat', 'start_lng']].mean().reset_index()
feb_data = feb_data.merge(station_coords, on='start_station_name', suffixes=('', '_standardized'))
```

```python
# Convert Data Types
jan_data['started_at'] = pd.to_datetime(jan_data['started_at'])
jan_data['ended_at'] = pd.to_datetime(jan_data['ended_at'])
feb_data['started_at'] = pd.to_datetime(feb_data['started_at'])
feb_data['ended_at'] = pd.to_datetime(feb_data['ended_at'])
```

```python
# Add Temporal columns for analysis 
jan_data['day'] = jan_data['started_at'].dt.day
jan_data['month'] = jan_data['started_at'].dt.month
jan_data['hour'] = jan_data['started_at'].dt.hour
jan_data['weekday'] = jan_data['started_at'].dt.dayofweek
jan_data['end_day'] = jan_data['ended_at'].dt.day
jan_data['end_month'] = jan_data['ended_at'].dt.month
jan_data['end_hour'] = jan_data['ended_at'].dt.hour
jan_data['end_weekday'] = jan_data['ended_at'].dt.dayofweek
```

```python
# Repeat for February
feb_data['day'] = feb_data['started_at'].dt.day
feb_data['month'] = feb_data['started_at'].dt.month
feb_data['hour'] = feb_data['started_at'].dt.hour
feb_data['weekday'] = feb_data['started_at'].dt.dayofweek
feb_data['end_day'] = feb_data['ended_at'].dt.day
feb_data['end_month'] = feb_data['ended_at'].dt.month
feb_data['end_hour'] = feb_data['ended_at'].dt.hour
feb_data['end_weekday'] = feb_data['ended_at'].dt.dayofweek
```

```python
# Calculate trip duration 
# Calculate trip duration in minutes
jan_data['trip_duration_minutes'] = ((jan_data['ended_at'] - jan_data['started_at']).dt.total_seconds() / 60).round()

# Verify the new column
jan_data[['started_at', 'ended_at', 'trip_duration_minutes']].head()
```

```python
# Calculate trip duration 
# Calculate trip duration in minutes
feb_data['trip_duration_minutes'] = ((feb_data['ended_at'] - feb_data['started_at']).dt.total_seconds() / 60).round()

# Verify the new column
feb_data[['started_at', 'ended_at', 'trip_duration_minutes']].head()
```

```python
# 1. Total number of rides per station
station_rides = winter_data.groupby('start_station_name').size().reset_index(name='total_rides')

# 2. Station Rank based on number of rides
station_rides['station_rank'] = station_rides['total_rides'].rank(ascending=False, method='dense').astype(int)

# 3. Average trip duration (by start station)
avg_trip_duration = winter_data.groupby('start_station_name')['trip_duration_minutes'].mean().reset_index(name='avg_trip_duration_minutes')

# 4. Most common end station for each start station
most_common_end_station = winter_data.groupby('start_station_name')['end_station_name'] \
    .agg(lambda x: x.value_counts().idxmax()).reset_index(name='most_common_end_station')

# Merge these calculated columns back into the main DataFrame
winter_data = winter_data.merge(station_rides, on='start_station_name', how='left')
winter_data = winter_data.merge(avg_trip_duration, on='start_station_name', how='left')
winter_data = winter_data.merge(most_common_end_station, on='start_station_name', how='left')
```
