```python
import pandas as pd
```

```python
# Load the CSV files for January and February 2023 
jan_data = pd.read_csv('Resources/202401-citibike-tripdata.csv')
feb_data = pd.read_csv('Resources/202402-citibike-tripdata.csv')
```

```python
# Check for missing values 
jan_data.isnull().sum()
feb_data.isnull().sum()
```

```python
# Calculate the total number of rows in each dataset
total_rows_jan = len(jan_data)
total_rows_feb = len(feb_data)

# Calculate the number of rows with missing data in each dataset
missing_rows_jan = jan_data[jan_data.isnull().any(axis=1)].shape[0]
missing_rows_feb = feb_data[feb_data.isnull().any(axis=1)].shape[0]

# Calculate the percentage of missing data
percent_missing_jan = (missing_rows_jan / total_rows_jan) * 100
percent_missing_feb = (missing_rows_feb / total_rows_feb) * 100

print(f"Percentage of rows with missing data in January dataset: {percent_missing_jan:.2f}%")
print(f"Percentage of rows with missing data in February dataset: {percent_missing_feb:.2f}%")
```

```python
# Remove all rows with missing values 
## Remove rows with missing values in January dataset
jan_data_cleaned = jan_data.dropna()

## Remove rows with missing values in February dataset
feb_data_cleaned = feb_data.dropna()
```

```python
# Check new row counts 

print(f"Original January dataset rows: {len(jan_data)}")
print(f"Cleaned January dataset rows: {len(jan_data_cleaned)}")
print(f"Original February dataset rows: {len(feb_data)}")
print(f"Cleaned February dataset rows: {len(feb_data_cleaned)}")
```

```python
# Combine January and February data
winter_data = pd.concat([jan_data_cleaned, feb_data_cleaned], ignore_index=True)
```

```python
# Drop duplicate rows 
winter_data.drop_duplicates(inplace=True)
```

```python
winter_data.head()
```

```python
winter_data.info()
```

```python
# Convert date and time columns 
winter_data['started_at'] = pd.to_datetime(winter_data['started_at'])
winter_data['ended_at'] = pd.to_datetime(winter_data['ended_at'])
```

```python
# # Convert Station ID to string
# winter_data['start_station_id'] = winter_data['start_station_id'].astype(str)
# winter_data['end_station_id'] = winter_data['end_station_id'].astype(str)
```

```python
# Confirm datatypes
winter_data.info()
```

```python
# Add Temporal columns for analysis 
winter_data['day'] = winter_data['started_at'].dt.day
winter_data['month'] = winter_data['started_at'].dt.month
winter_data['hour'] = winter_data['started_at'].dt.hour
winter_data['weekday'] = winter_data['started_at'].dt.dayofweek
winter_data['end_day'] = winter_data['ended_at'].dt.day
winter_data['end_month'] = winter_data['ended_at'].dt.month
winter_data['end_hour'] = winter_data['ended_at'].dt.hour
winter_data['end_weekday'] = winter_data['ended_at'].dt.dayofweek
```

```python
# Calculate trip duration 
# Calculate trip duration in minutes
winter_data['trip_duration_minutes'] = ((winter_data['ended_at'] - winter_data['started_at']).dt.total_seconds() / 60).round()

# Verify the new column
winter_data[['started_at', 'ended_at', 'trip_duration_minutes']].head()
```

```python
# Convert station names to uppercase to ensure consistency
winter_data['start_station_name'] = winter_data['start_station_name'].str.upper()
winter_data['end_station_name'] = winter_data['end_station_name'].str.upper()
```

```python
# Identify number of unique station names 
unique_start_names = winter_data['start_station_name'].nunique()
unique_end_names = winter_data['end_station_name'].nunique()
unique_start_ids = winter_data['start_station_id'].nunique()
unique_end_ids = winter_data['end_station_id'].nunique()

print(f"Unique start station IDs: {unique_start_ids}")
print(f"Unique end station IDs: {unique_end_ids}")
print(f"Unique start station names: {unique_start_names}")
print(f"Unique end station names: {unique_end_names}")
```

```python
# Find station names with "Ave" vs "Avenue", "St" vs "Street", etc.
ave_names = winter_data[winter_data['start_station_name'].str.contains(r'\bAVENUE\b', case=False, na=False)]['start_station_name'].unique()
st_names = winter_data[winter_data['start_station_name'].str.contains(r'\bSTREET\b', case=False, na=False)]['start_station_name'].unique()
rd_names = winter_data[winter_data['start_station_name'].str.contains(r'\bROAD\b', case=False, na=False)]['start_station_name'].unique()
dr_names = winter_data[winter_data['start_station_name'].str.contains(r'\bDRIVE\b', case=False, na=False)]['start_station_name'].unique()
period_names = winter_data[winter_data['start_station_name'].str.contains(r'\.', na=False)]['start_station_name'].unique()

print("Stations with 'Ave':", ave_names)
print(len(ave_names))
print("Stations with 'St':", st_names)
print(len(st_names))
print("Stations with 'Road:'", rd_names)
print(len(rd_names))
print("Stations with '.'", period_names)
print(len(period_names))
print("Stations with '.'", dr_names)
print(len(dr_names))
```

```python
# Standardize common abbreviations
winter_data['start_station_name'] = winter_data['start_station_name'].str.replace(r'\bAVENUE\b', 'AVE', regex=True)
winter_data['start_station_name'] = winter_data['start_station_name'].str.replace(r'\bSTREET\b', 'ST', regex=True)
winter_data['start_station_name'] = winter_data['start_station_name'].str.replace(r'\bROAD\b', 'RD', regex=True)
winter_data['start_station_name'] = winter_data['start_station_name'].str.replace(r'\bPLACE\b', 'PL', regex=True)

winter_data['end_station_name'] = winter_data['end_station_name'].str.replace(r'\bAVENUE\b', 'AVE', regex=True)
winter_data['end_station_name'] = winter_data['end_station_name'].str.replace(r'\bSTREET\b', 'ST', regex=True)
winter_data['end_station_name'] = winter_data['end_station_name'].str.replace(r'\bROAD\b', 'RD', regex=True)
winter_data['end_station_name'] = winter_data['end_station_name'].str.replace(r'\bPLACE\b', 'PL', regex=True)
```

```python
# Remove periods from station names
winter_data['start_station_name'] = winter_data['start_station_name'].str.replace('.', '', regex=False)
winter_data['end_station_name'] = winter_data['end_station_name'].str.replace('.', '', regex=False)
```

```python
# Identify number of unique station names 
unique_start_names = winter_data['start_station_name'].nunique()
unique_end_names = winter_data['end_station_name'].nunique()
unique_start_ids = winter_data['start_station_id'].nunique()
unique_end_ids = winter_data['end_station_id'].nunique()

print(f"Unique start station IDs: {unique_start_ids}")
print(f"Unique end station IDs: {unique_end_ids}")
print(f"Unique start station names: {unique_start_names}")
print(f"Unique end station names: {unique_end_names}")
```

Standarize Station Coordinates 

```python
# Step 1: Calculate the mean coordinates for each start station name
# Group by start_station_name and calculate the average coordinates
station_coords = winter_data.groupby('start_station_name')[['start_lat', 'start_lng']].mean()
station_coords = station_coords.rename(columns={'start_lat': 'standardized_start_lat', 'start_lng': 'standardized_start_lng'})

# Display the standardized coordinates
print(station_coords.head())

# Step 2: Map these standardized coordinates back to the main dataframe
# Map the standardized coordinates to the 'start_station_name' and 'end_station_name'
winter_data = winter_data.merge(station_coords, how='left', left_on='start_station_name', right_index=True)

# Repeat the process to get the standardized coordinates for end stations
end_station_coords = winter_data.groupby('end_station_name')[['end_lat', 'end_lng']].mean()
end_station_coords = end_station_coords.rename(columns={'end_lat': 'standardized_end_lat', 'end_lng': 'standardized_end_lng'})

# Map the standardized coordinates to the main dataframe
winter_data = winter_data.merge(end_station_coords, how='left', left_on='end_station_name', right_index=True)

# Display a preview to confirm that standardized coordinates are added
print(winter_data[['start_station_name', 'standardized_start_lat', 'standardized_start_lng', 
                   'end_station_name', 'standardized_end_lat', 'standardized_end_lng']].head())
```

```python
# Check for any NaN values in the standardized columns
missing_start_coords = winter_data[['standardized_start_lat', 'standardized_start_lng']].isna().sum()
missing_end_coords = winter_data[['standardized_end_lat', 'standardized_end_lng']].isna().sum()

print("Missing standardized start coordinates:", missing_start_coords)
print("Missing standardized end coordinates:", missing_end_coords)
```

```python
# Replace original coordinates with standardized coordinates for start and end stations
winter_data['start_lat'] = winter_data['standardized_start_lat']
winter_data['start_lng'] = winter_data['standardized_start_lng']
winter_data['end_lat'] = winter_data['standardized_end_lat']
winter_data['end_lng'] = winter_data['standardized_end_lng']

# Drop the standardized columns 
winter_data = winter_data.drop(columns=['standardized_start_lat', 'standardized_start_lng', 
                                        'standardized_end_lat', 'standardized_end_lng'])

# Confirm the changes 
winter_data.head()
```

Total Number of Rides per Start Station

```python
# Total number of rides per start station
rides_per_start_station = winter_data.groupby('start_station_name')['ride_id'].count().reset_index()
rides_per_start_station.columns = ['start_station_name', 'total_rides_start']
```

Start Station Rank Based on Total Number of Rides

```python
# Start station rank based on total rides
rides_per_start_station['start_station_rank'] = rides_per_start_station['total_rides_start'].rank(
    method='dense', ascending=False
).astype(int)
```

 Total Number of Rides per End Station

```python
# Total number of rides per end station
rides_per_end_station = winter_data.groupby('end_station_name')['ride_id'].count().reset_index()
rides_per_end_station.columns = ['end_station_name', 'total_rides_end']
```

```python
End Station Rank Based on Total Number of Rides
```

```python
# End station rank based on total rides
rides_per_end_station['end_station_rank'] = rides_per_end_station['total_rides_end'].rank(
    method='dense', ascending=False
).astype(int)
```

Most Popular End Station for Each Start Station

```python
# Most popular end station for each start station
popular_end_station = winter_data.groupby(['start_station_name', 'end_station_name'])['ride_id'].count()
popular_end_station = popular_end_station.reset_index().rename(columns={'ride_id': 'ride_count'})
popular_end_station = popular_end_station.loc[popular_end_station.groupby('start_station_name')['ride_count'].idxmax()]
popular_end_station = popular_end_station[['start_station_name', 'end_station_name']]
popular_end_station.columns = ['start_station_name', 'most_popular_end_station']
```

Most Popular Day of the Week per Start Station

```python
# Most popular day of the week per start station
popular_day_start = winter_data.groupby(['start_station_name', 'weekday'])['ride_id'].count()
popular_day_start = popular_day_start.reset_index().rename(columns={'ride_id': 'ride_count'})
popular_day_start = popular_day_start.loc[popular_day_start.groupby('start_station_name')['ride_count'].idxmax()]
popular_day_start = popular_day_start[['start_station_name', 'weekday']]
popular_day_start.columns = ['start_station_name', 'most_popular_day_start']
```

 Most Popular Day of the Week per End Station

```python
# Most popular day of the week per end station
popular_day_end = winter_data.groupby(['end_station_name', 'weekday'])['ride_id'].count()
popular_day_end = popular_day_end.reset_index().rename(columns={'ride_id': 'ride_count'})
popular_day_end = popular_day_end.loc[popular_day_end.groupby('end_station_name')['ride_count'].idxmax()]
popular_day_end = popular_day_end[['end_station_name', 'weekday']]
popular_day_end.columns = ['end_station_name', 'most_popular_day_end']
```

 Most Popular Hour of the Day per Start Station

```python
# Most popular hour of the day per start station
popular_hour_start = winter_data.groupby(['start_station_name', 'hour'])['ride_id'].count()
popular_hour_start = popular_hour_start.reset_index().rename(columns={'ride_id': 'ride_count'})
popular_hour_start = popular_hour_start.loc[popular_hour_start.groupby('start_station_name')['ride_count'].idxmax()]
popular_hour_start = popular_hour_start[['start_station_name', 'hour']]
popular_hour_start.columns = ['start_station_name', 'most_popular_hour_start']
```

Most Popular Hour of the Day per End Station

```python
# Most popular hour of the day per end station
popular_hour_end = winter_data.groupby(['end_station_name', 'hour'])['ride_id'].count()
popular_hour_end = popular_hour_end.reset_index().rename(columns={'ride_id': 'ride_count'})
popular_hour_end = popular_hour_end.loc[popular_hour_end.groupby('end_station_name')['ride_count'].idxmax()]
popular_hour_end = popular_hour_end[['end_station_name', 'hour']]
popular_hour_end.columns = ['end_station_name', 'most_popular_hour_end']
```

 Average Trip Duration for Each Start Station

```python
# Average trip duration per start station
avg_duration_start_station = winter_data.groupby('start_station_name')['trip_duration_minutes'].mean().reset_index()
avg_duration_start_station.columns = ['start_station_name', 'average_trip_duration_start']
avg_duration_start_station['average_trip_duration_start'] = avg_duration_start_station['average_trip_duration_start'].round(2)
```

Station Breakdown

```python
# Merge all metrics into a single DataFrame
user_type_analysis = rides_per_start_station.merge(
    rides_per_end_station, left_on='start_station_name', right_on='end_station_name', how='outer'
).merge(
    popular_end_station, on='start_station_name', how='outer'
).merge(
    popular_day_start, on='start_station_name', how='outer'
).merge(
    popular_day_end, left_on='end_station_name', right_on='end_station_name', how='outer'
).merge(
    popular_hour_start, on='start_station_name', how='outer'
).merge(
    popular_hour_end, left_on='end_station_name', right_on='end_station_name', how='outer'
).merge(
    avg_duration_start_station, on='start_station_name', how='outer'
)

# Fill any NaN values (optional)
user_type_analysis = user_type_analysis.fillna('N/A')

# Display the result using standard Pandas display
user_type_analysis.head()
```

```python
# Replace 'N/A' with 0 in relevant columns
user_type_analysis[['total_rides_start', 'start_station_rank', 'total_rides_end', 'end_station_rank', 'average_trip_duration_start']] = (
    user_type_analysis[['total_rides_start', 'start_station_rank', 'total_rides_end', 'end_station_rank', 'average_trip_duration_start']]
    .replace("N/A", 0)
    .astype(int)
)
```

```python
# Convert rank and ride counts to integers
user_type_analysis['total_rides_start'] = user_type_analysis['total_rides_start'].fillna(0).astype(int)
user_type_analysis['start_station_rank'] = user_type_analysis['start_station_rank'].fillna(0).astype(int)
user_type_analysis['total_rides_end'] = user_type_analysis['total_rides_end'].fillna(0).astype(int)
user_type_analysis['end_station_rank'] = user_type_analysis['end_station_rank'].fillna(0).astype(int)

# Convert trip duration to integer if you want to round it to remove decimals
user_type_analysis['average_trip_duration_start'] = user_type_analysis['average_trip_duration_start'].round(0).astype(int)
```

```python
# Map numbers to day names
day_mapping = {0: "Monday", 1: "Tuesday", 2: "Wednesday", 3: "Thursday", 4: "Friday", 5: "Saturday", 6: "Sunday"}
user_type_analysis['most_popular_day_start'] = user_type_analysis['most_popular_day_start'].replace(day_mapping)
user_type_analysis['most_popular_day_end'] = user_type_analysis['most_popular_day_end'].replace(day_mapping)
```

```python
# Step 1: Convert most_popular_hour columns to 12-hour format with AM/PM
user_type_analysis['most_popular_hour_start'] = user_type_analysis['most_popular_hour_start'].apply(
    lambda x: f"{int(x % 12 if x % 12 else 12)} {'AM' if x < 12 else 'PM'}" if x != 'N/A' else 'N/A'
)
user_type_analysis['most_popular_hour_end'] = user_type_analysis['most_popular_hour_end'].apply(
    lambda x: f"{int(x % 12 if x % 12 else 12)} {'AM' if x < 12 else 'PM'}" if x != 'N/A' else 'N/A'
)

# Step 2: Remove decimals from all numeric columns by converting to integer
numeric_columns = ['total_rides_start', 'start_station_rank', 'total_rides_end', 'end_station_rank', 'average_trip_duration_start']
user_type_analysis[numeric_columns] = user_type_analysis[numeric_columns].replace("N/A", 0).astype(int)
```

```python
import numpy as np

# Add 1 to start and end station ranks to make them start from 1
user_type_analysis['start_station_rank'] = user_type_analysis['start_station_rank'] + 1
user_type_analysis['end_station_rank'] = user_type_analysis['end_station_rank'] + 1

# Set rank to NA for stations with 0 rides
user_type_analysis.loc[user_type_analysis['total_rides_start'] == 0, 'start_station_rank'] = np.nan
user_type_analysis.loc[user_type_analysis['total_rides_end'] == 0, 'end_station_rank'] = np.nan
```
